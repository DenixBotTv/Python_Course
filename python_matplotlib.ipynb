## plt.subplots() in matplotlib
This function creates a **figure** and one or more **subplots** (axes objects) at once. It's a very common way to build more flexible and professional plots.

````python
x = [1, 2, 3]
y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Simple Plot")
plt.show()
````

If you need to plot individual points, you can combine *subplots* with *scatter*

````python
x = [1, 2, 3, 4]
y = [10, 15, 13, 17]

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Scatter Plot")
plt.show()
````

## Random walk
Let's create a random walk

````python
step_option = [1,-1]
step_choice = choice(step_option)

walk = []
walk.append(step_choice)
for step in range(1,1000):
    next_step = walk[step-1] + choice(step_option)
    walk.append(next_step)

plt.plot(walk)
plt.show()
````

We can refactor our code using functions

````python
def rand_walk(step_num):
    walk = []
    step_choice = choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice_2 = choice([1,-1])
        next_step = walk[i-1] + step_choice_2
        walk.append(next_step)
    return walk  

def plot_walk(walk):
    plt.plot(walk)
    plt.xlabel('Number of Steps')
    plt.ylabel('Distance from Origin')
    plt.title('Our Random Walk')
    plt.show()

random_w = rand_walk(100)
plot_walk(rand_walk(1000))
````

And finally we can create a new function that can draw multiple random walks

````python
def plot_walks(num,walk_steps):
    
    lab_list = list(range(1,num+1))
    for i in range(0,num):
        
        x = list(range(1,walk_steps+1))
    
        plt.plot(x,rand_walk(walk_steps),label = 'Plot Number' + ' ' + str(lab_list[i]))
        plt.xlabel('Number of Steps')
        plt.ylabel('Distance from Origin')
        plt.title('Our Random Walk')
        plt.legend(loc='lower left')
    plt.show()

plot_walks(5,100000)
````

You can do also with *subplots* and *scatter*

````python
def sc_rand_walk(step_num):
    walk = []
    step_choice = choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice_2 = choice([1,-1])
        next_step = walk[i-1] + step_choice_2
        walk.append(next_step)
    return walk

def sc_plot_walks(num, walk_steps):
    fx, ax = plt.subplots()
    for i in range(1,num):
        y = sc_rand_walk(walk_steps)
        x = range(1, len(y)+1)

        ax.scatter(x,y, label = 'Plot Number' + ' ' + str(i), s=10)
        ax.set_xlabel('Number of  Steps')
        ax.set_ylabel('Distance from Origin')
        ax.set_title('Our Random Walk')
        ax.legend(loc='lower left')
    plt.show()

sc_plot_walks(5,1000)
````

## SIerpinski Triangle
This is a very nice app that is based upon 3 transformations:
First Transformation 
\begin{equation}
x_{n+1} = 0.5x_n 
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}
Second Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 0.5
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n + 0.5
\end{equation}
Third Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 1
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}

Let's develop a Python script that generate some points according to one trasformation at time choosen randoly


````python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1,y1


transformations = [trans_1,trans_2,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0
 
for i in range(100):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
````

````python
plt.rc('figure',figsize=(16,16))

plt.scatter(a1,b1,s=5)
#plt.savefig('my_figure.png')
plt.show()
````
